## Binary Classification

```python
def sigmoid(X):
    
    Y = 1 / (1 + np.exp(-X))
    
    return Y    
```

```python
x_temp = np.random.rand(5,6)
y_temp = sigmoid(x_temp)
print(x_temp)
print(y_temp)
```

## Cost function

```python
def cost_fun(y,h):
    
    cost = -1/y.size * np.sum( np.dot(y.T,np.log(h)) + np.dot(1-y.T, np.log(1-h)))
    
    return cost
```

```python
def gradient(X):
    
    X = sigmoid(X)*(1-sigmoid(X))
    
    return X    
```

```python
x_temp = np.random.rand(5,6)

y_temp = [1.,0.,1.,1.,0.]
y_temp = np.array(y_temp)
y_temp = np.expand_dims(y_temp,axis = 1)

w_temp = np.random.rand(x_temp.shape[1],1)

y_hat = sigmoid(np.dot(x_temp,w_temp))

cost = cost_fun(y_temp,y_hat)
print("cost : {}".format(cost))

g = gradient(np.dot(x_temp,w_temp))
print(f"gradient : {g}")
```

## Gradient descent 
### Now we will define theta a 5*3 matrix which will give gradient for each parameter  for each class

Weight matrix for logistice regression will have shape = (no_of_features + 1, no_of_classes)

```python
def initialize_parameters(x,classes):
    
    n = x.shape[1]
    theta = np.random.rand(n, classes)*0.01
    
    return theta
```

```python
def predict_prob(x, theta):
    
    h = sigmoid(np.dot(x,theta))
    
    return h
```

```python
def accuracy(h, y):
    
    m = y.shape[0]
    h = np.argmax(h,axis=1)                          # max value in each row of h array of sigmoid function
    q = np.argmax(y,axis=1) == h                     # check how many output we got correct
    print(np.sum(q*100/m))
    
```

## Model

```python
def model(x, y, classes = 3, learning_rate = 1, iterations = 1000):
    
    m = x.shape[0]
    theta = initialize_parameters(x, classes)
    
    for i in range (iterations):
        h = predict_prob(x, theta)                          # sigmoid function update
        theta -= ((learning_rate)/m)*(np.dot((h-y).T , x).T)    # theta update
    
    accuracy(h,y)
    
    return theta
```
